```python
df['country'].fillna('unknown',inplace=True)
```

```python
countries=df['country'].value_counts().reset_index()
print(countries)
```

```python


fig = px.pie(
    countries,
    values='count',
    names='country',
    title='Countries'
)

fig.update_traces(
    textposition='inside',   # Move text inside slices
    textinfo='percent+label', # Show both percent and label
    pull=[0.02]*len(countries) # Slightly pull slices to improve visibility
)

fig.update_layout(
    width=1200,
    height=700,
    title_x=0.5,              # Center the title
    margin=dict(l=100, r=100, t=100, b=100),
    showlegend=True
)

fig.show()
```
